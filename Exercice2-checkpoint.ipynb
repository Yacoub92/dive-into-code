```python
# Problem 1

import matplotlib.pyplot as plt

def compute_sorori_shinzaemon(n_days=100):
    # Initialize variables
    grains = 1
    list_n_grains = []
    list_total_grains = []

    # Loop to calculate number of grains of rice on each day
    for i in range(n_days):
        list_n_grains.append(grains)
        list_total_grains.append(sum(list_n_grains))
        grains *= 2

    # Plot line graph
    plt.plot(range(1, n_days+1), list_n_grains, label='Grains per day')
    plt.plot(range(1, n_days+1), list_total_grains, label='Total grains')
    plt.xlabel('Days')
    plt.ylabel('Number of rice grains')
    plt.legend()
    plt.show()

    return list_n_grains[-1], list_total_grains[-1]

# Call the function with default argument (100 days)
grains_on_100th_day, total_grains = compute_sorori_shinzaemon()
print("Number of grains on 100th day:", grains_on_100th_day)
```

```python
# Problem 2 

def compute_sorori_shinzaemon(n_days=100):
    """
    A function that calculates the number of rice grains per day in the rice anecdote of Shinzaemon Sorori.

    Parameters
    ----------
    n_days : int
        Number of days to receive rice (default: 100)

    Returns
    -------
    list_n_grains : list
        List of the number of rice grains received per day (list length equals n_days)
    list_total_grains : list
        List of the total number of rice grains received by a certain day (list length is equal to n_days)
    """

    # Initializing variables
    n_grains = 1  # Starting with 1 grain of rice on the first day
    list_n_grains = []  # Empty list to store number of grains received each day
    list_total_grains = []  # Empty list to store total grains received by each day

    # the number of days
    for day in range(1, n_days+1):

        # Add the number of grains received to the list for the day
        list_n_grains.append(n_grains)

        # Add the number of grains received to the total grains list
        total_grains = sum(list_n_grains)
        list_total_grains.append(total_grains)

        # Double the number of grains for the next day
        n_grains *= 2

    return list_n_grains, list_total_grains
```

```python
# Calculate rice grains for 30 days
list_n_grains, list_total_grains = compute_sorori_shinzaemon(n_days=30)

# Print the total rice grains received for 30 days
print(f"Total rice grains received for 30 days: {list_total_grains[-1]}")
```
