```python
import numpy as np
import matplotlib.pyplot as plt

# Set the seed value for reproducibility
np.random.seed(0)

# Define the mean and covariance matrix
mean = (-3, 0)
cov = [[1.0, 0.8], [0.8, 1.0]]

# Generate the random numbers with a 2D normal distribution
data = np.random.multivariate_normal(mean, cov, 500)

# Plot the data in histograms
plt.subplot(1, 2, 1)
plt.hist(data[:, 0], bins=20, range=(-6, 0))
plt.title('Histogram of X-axis')
plt.xlabel('X-axis')
plt.ylabel('Frequency')

plt.subplot(1, 2, 2)
plt.hist(data[:, 1], bins=20, range=(-6, 6))
plt.title('Histogram of Y-axis')
plt.xlabel('Y-axis')
plt.ylabel('Frequency')

plt.subplots_adjust(wspace=0.4)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Set the seed value for reproducibility
np.random.seed(0)

# Define the mean and covariance matrices for both sets of data
mean1 = (-3, 0)
cov1 = [[1.0, 0.8], [0.8, 1.0]]
mean2 = (0, -3)
cov2 = [[1.0, 0.8], [0.8, 1.0]]

# Generate the random numbers with a 2D normal distribution for both sets of data
data1 = np.random.multivariate_normal(mean1, cov1, 500)
data2 = np.random.multivariate_normal(mean2, cov2, 500)

# Combine the data into one array
data = np.concatenate((data1, data2), axis=0)

# Create a scatter plot of the combined data
plt.scatter(data1[:, 0], data1[:, 1], label='0')
plt.scatter(data2[:, 0], data2[:, 1], label='1')
plt.title('Scatter plot of combined data')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.show()
```
