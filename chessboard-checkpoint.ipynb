```python
# Problem 2

import numpy as np

def wheat_chessboard(n, m):
    """
    Returns an ndarray of wheat counts on an nxm chessboard.
    """
    # Create a list of wheat grain values for an nxm chessboard
    wheat_grains = [1]
    for _ in range(n*m - 1):
        wheat_grains.append(2*wheat_grains[-1])
        
    # Create an ndarray from the list and reshape it to an nxm ndarray
    chessboard = np.array(wheat_grains).reshape((n, m))
    
    return chessboard

# Calculate the number of wheat grains on an 8x8 chessboard
chessboard_8x8 = wheat_chessboard(8, 8)
total_wheat = np.sum(chessboard_8x8)

# Print the number of wheat grains
print("Total number of wheat grains on an 8x8 chessboard: {}".format(total_wheat))
```

```python
# Problem 3

import numpy as np
import matplotlib.pyplot as plt

def wheat_chessboard(n, m):
    """
    Returns an ndarray of wheat counts on an nxm chessboard.
    """
    # Create a list of wheat grain values for an nxm chessboard
    wheat_grains = [1]
    for _ in range(n*m - 1):
        wheat_grains.append(2*wheat_grains[-1])
        
    # Create an ndarray from the list and reshape it to an nxm ndarray
    chessboard = np.array(wheat_grains).reshape((n, m))
    
    return chessboard

# Calculate the number of wheat grains on an 8x8 chessboard
chessboard_8x8 = wheat_chessboard(8, 8)
total_wheat = np.sum(chessboard_8x8)

# Print the total number of wheat grains
print("Total number of wheat grains on an 8x8 chessboard: {}".format(total_wheat))

# Calculate the average wheat grains per column
avg_wheat_per_column = np.mean(chessboard_8x8, axis=0)

# Visualize the average wheat grains per column in a bar chart
plt.xlabel("Column")
plt.ylabel("Number")
plt.title("Average number of wheat grains per column on an 8x8 chessboard")
plt.bar(np.arange(1, 9), avg_wheat_per_column)
plt.show()
```

```python
# Problem 5

import numpy as np

# create the chessboard
n_rows = 8
n_cols = 8
small_board_list = [1]
for _ in range(n_rows * n_cols - 1):
    small_board_list.append(2 * small_board_list[-1])

chessboard = np.array(small_board_list).reshape(n_rows, n_cols)

# calculate the number of wheat in the first half of the chessboard
first_half = chessboard[:4, :].sum()

# calculate the number of wheat in the second half of the chessboard
second_half = chessboard[4:, :].sum()

# calculate the product of the two halves
result = first_half * second_half

print("The number of wheat in the first half is:", first_half)
print("The number of wheat in the second half is:", second_half)
print("The second half is {} times the first half.".format(second_half // first_half))
print("The product of the two halves is:", result)
```

```python
# Problem 6

import numpy as np

def calc_wheat_on_board_append(n, m):
    board_ndarray = np.array([1])
    for _ in range(n * m - 1):
        board_ndarray = np.append(board_ndarray, 2*board_ndarray[-1])
    return board_ndarray.reshape(n, m)

import numpy as np

def calc_wheat_on_board_broadcast(n, m):
    indices_of_squares = np.arange(n * m, dtype=np.uint64)
    board_ndarray = 2**indices_of_squares
    return board_ndarray.reshape(n, m)

board1 = calc_wheat_on_board_append(8, 8)
board2 = calc_wheat_on_board_broadcast(8, 8)

print("Number of wheat on 8x8 chessboard using np.append():\n", board1)
print("Number of wheat on 8x8 chessboard using broadcasting:\n", board2)
print("Are the two arrays equal? ", np.array_equal(board1, board2))
```
