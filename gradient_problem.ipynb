```python
# Problem 3

import numpy as np

# Define the function for y
def f(x):
    return 0.5 * x + 1

# Create a numpy ndarray representing x at 0.1 intervals from -50 to 50
x = np.arange(-50, 50.1, 0.1)

# Calculate the exact gradient at each x-coordinate using the limit (derivative)
h = 0.0001  # small value for approximating derivative using limit
gradient_exact = np.zeros_like(x[:-1])  # initialize gradient array
for i in range(len(x) - 1):
    # Approximate the derivative at x[i] using the limit (derivative)
    numerator = f(x[i] + h) - f(x[i])
    denominator = h
    gradient_exact[i] = numerator / denominator

# Print the resulting array of exact gradients
print(gradient_exact)
```

```python
# Problem 4

import numpy as np
import matplotlib.pyplot as plt

# Define the function for y
def f(x):
    return 0.5 * x + 1

# Create a numpy ndarray representing x at 0.1 intervals from -50 to 50
x = np.arange(-50, 50.1, 0.1)

# Calculate y for each element of x using the provided formula
y = f(x)

# Calculate the change in y and change in x between adjacent elements
delta_y = np.diff(y)
delta_x = np.diff(x)

# Calculate the gradient by dividing change in y by change in x
gradient = delta_y / delta_x

# Plot the linear function
plt.figure(figsize=(10, 5))
plt.plot(x, y)
plt.title("Linear Function")
plt.xlabel("x")
plt.ylabel("y")
plt.grid()

# Plot x vs gradient
plt.figure(figsize=(10, 5))
plt.plot(x[:-1], gradient)
plt.title("Gradient vs x")
plt.xlabel("x")
plt.ylabel("Gradient")
plt.grid()

plt.show()
```

```python
# Problem 5

import numpy as np
import matplotlib.pyplot as plt

def compute_gradient(function, x_range=(-50, 50.1, 0.1)):
    
    x = np.arange(*x_range)
    y = function(x)
    delta_y = np.diff(y)
    delta_x = np.diff(x)
    gradient = delta_y / delta_x
    array_xy = np.column_stack((x, y))
    return array_xy, gradient

def function1(x):
    return 0.5 * x + 1

def function2(x):
    return 2 * x**2 + 2 * x

def function3(x):
    return np.sin(x/2)

array_xy1, gradient1 = compute_gradient(function1)
array_xy2, gradient2 = compute_gradient(function2)
array_xy3, gradient3 = compute_gradient(function3, x_range=(0, 50.1, 0.1))

fig, axs = plt.subplots(2, 2, figsize=(12, 8))
fig.tight_layout(pad=3.0)

axs[0, 0].plot(array_xy1[:, 0], array_xy1[:, 1])
axs[0, 0].set_title('Linear Function')
axs[0, 0].set_xlabel('x')
axs[0, 0].set_ylabel('y')
axs[0, 0].grid()

axs[1, 0].plot(array_xy1[:-1, 0], gradient1)
axs[1, 0].set_title('Gradient vs x')
axs[1, 0].set_xlabel('x')
axs[1, 0].set_ylabel('Gradient')
axs[1, 0].grid()

axs[0, 1].plot(array_xy2[:, 0], array_xy2[:, 1])
axs[0, 1].set_title('Quadratic Function')
axs[0, 1].set_xlabel('x')
axs[0, 1].set_ylabel('y')
axs[0, 1].grid()

axs[1, 1].plot(array_xy2[:-1, 0], gradient2)
axs[1, 1].set_title('Gradient vs x')
axs[1, 1].set_xlabel('x')
axs[1, 1].set_ylabel('Gradient')
axs[1, 1].grid()

axs[0, 2].plot(array_xy3[:, 0], array_xy3[:, 1])
axs[0, 2].set_title('Sine Function')
axs[0, 2].set_xlabel('x')
axs[0, 2].set_ylabel('y')
axs[0, 2].grid()

axs[1, 2].plot(array_xy3[:-1, 0], gradient3)
axs
```

```python
# Problem 6

import numpy as np
import matplotlib.pyplot as plt

# Define the compute_gradient function
def compute_gradient(function, x_range=(-50, 50.1, 0.1)):
    
    x = np.arange(*x_range)
    y = function(x)
    gradient = np.diff(y) / np.diff(x)
    array_xy = np.concatenate([x.reshape(-1,1), y.reshape(-1,1)], axis=1)
    return array_xy, gradient

# Define the function to calculate y
def function5(array_x):
    return np.sin(array_x**0.5)

# Calculate the gradient using compute_gradient function
array_xy, gradient = compute_gradient(function5)

# Find the minimum y and its index
min_y = np.min(array_xy[:, 1])
min_y_idx = np.argmin(array_xy[:, 1])

# Find the slope before and after the minimum y index
slope_before = gradient[min_y_idx - 1]
slope_after = gradient[min_y_idx]

# Print the results
print(f"The minimum value of y is {min_y:.4f} at x = {array_xy[min_y_idx, 0]:.4f}")
print(f"The slope before the minimum y is {slope_before:.4f}")
print(f"The slope after the minimum y is {slope_after:.4f}")
```
